```
# Imports 
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style('darkgrid')
sns.set(font_scale=1.5)

# Read in dataset
df = pd.read_csv('Data1.csv')

df.describe()
```
